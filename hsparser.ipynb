```python
import re

sample_string = """\
1000-1000;\
1000-2000,2010,2020;\
1000-3000~3002,3010~3012;\
1000-4000,4010~4011;\
1000-5000~5002,5010;\
2000~2002\
"""

def parseStringToCodes(string):
    
    string = re.sub(' ', '', string)
    pos = re.sub('減\((.*?)\)', '', string)
    neg = ';'.join(re.findall('減\((.*?)\)', string))
    
    def parseString(string):
        
        # Handle "~"
        def expandRange(matchObject):
            return ','.join([str(x) for x in range(int(matchObject.group(1)), int(matchObject.group(2))+1)])                      

        codes = re.sub('(\d+)~(\d+)', expandRange, string)

        # Handle ","
        def appendPrefix(matchObject):
            prefix = matchObject.group(1) + '-'
            ls = [prefix + s for s in matchObject.group(2).split(',')]
            return ';'.join(ls)

        codes = re.sub('(\d+)-([^;]*)', appendPrefix, codes)

        # Handle "-"
        codes = re.sub('-', '', codes)

        # Create list
        return re.split(';|,', codes)

    return dict(pos=parseString(pos), neg=parseString(neg))

parseStringToCodes(sample_string)
```
